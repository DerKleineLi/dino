```python
import matplotlib.pyplot as plt
import experiment.logs as logs
from pathlib import Path


log = logs.read_logs(
    Path("/mnt/raid/hli/models/dino"),
    {
        "train_subset": "train_subset/log.txt",
        "train_no_shed": "train_no_shed/log.txt",
        "train_drop": "train_drop/log.txt",
        "train_drop_no_shed": "train_drop_no_shed/log.txt",
        "train_no_head": "train_no_head/log.txt",
        "overfit_drop_no_shed": "overfit_drop_no_shed/log.txt",
        "train_drop_no_aug": "train_drop_no_aug/log.txt",
    },
)
```

```python
import pandas as pd

log_dict = {key: pd.DataFrame(value) for key, value in log.items()}
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for key, df in log_dict.items():
    df.plot("epoch", "train_loss", ax=ax, label=key)
# ax.set_xlim([0, 100])
# ax.set_ylim([5,12])
```
